## UV

### Installation

In the sample application, we are using UV, which is a very popular and fast package manager for python. 

The installation of `uv` should be pretty straight-forward: https://docs.astral.sh/uv/
The easiest way is to open up a new terminal in Visual Studio and perform the commands in there.

You probably need to start a new terminal session after installing uv. 

After installation, please open up a terminal in VS Code and execute `uv --version` to verify your installation

If you have uv already installed, it doesn't hurt to update it to its latest version with `uv self update`

### Creating the virtual environment and install the project dependencies

Now that you have `uv` installed, its time to run `uv sync` in your terminal. 

This does the following: 
1. A new virtual environment is created - You now have a .venv folder in your project root. This Virtual Environment is a small isolated Python setup just for this project.
2. All project dependencies that are defined in `pyproject.toml` are installed in that new virtual environment. 

Now you should be able to run cells a Jupyter Notebook (like this file). 

For each notebook you need to first define the kernel on which you want to run. 

Select "Python Environments..." and then the one that references the .venv folder of our project (It is usually highlighted with a start in VS Code)

After you have selected the kernel, try to execute the next cell. It should sucessfully print out the result of the complex calculation. 

```python
result = 1 + 1

print(result)
```

If you have problems with executing the cell, maybe ChatGPT or another AI tool can help you. If you are still stuck, don't hesitate to shoot me a question via udemy!

Maybe you wonder, why we didn't have to install Pyhton on or machine... The reason for this is that `uv` itself installs a verions of python when it creates the virtual environment. This python version is located under .venv/bin/python.

The version that should be used for the virtual environment can be defined in the .python-version file.
