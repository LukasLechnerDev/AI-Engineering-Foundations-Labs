# Your First LLM call to local oss model!

1. Download and install ollama: https://ollama.com/
2. Verify installation by running `ollama` in the terminal
3. Download a model from: https://ollama.com/search
    3.1 Recommendation: Gemma3:4b from Google - run `ollama pull gemma3:4b`
4. Interact with the model by executing `ollama` in the terminal and select `Run a model (gemma3:4b)`
5. OpenAI Python SDK can be used since Ollama is OpenAI-compatible: https://docs.ollama.com/api/openai-compatibility see below 👇


```python
from openai import OpenAI
```

```python
# Ollama's default port is 11434
OLLAMA_BASE_URL = "http://localhost:11434/v1"

client = OpenAI(
    base_url=OLLAMA_BASE_URL, 
    api_key='ollama' # required but ignored
)

prompt = "Who are you?"

response = client.responses.create(
    model="gemma3:4b", 
    input=prompt
)

print(response.output_text)
```
