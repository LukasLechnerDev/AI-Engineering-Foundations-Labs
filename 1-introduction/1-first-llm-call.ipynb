# Your First LLM call

We will now make our first call to the OpenAI API!
See the OpenAI Developer Quickstart for reference: https://developers.openai.com/api/docs/quickstart

But first, we need to define our OpenAI API key in the `.env` file. 
The project already contains a `.env.example` file, that can be used as a reference about how the environment variables should be named:


- Perform `cp .env.example .env` in your terminal to create your `.env` file from the example
- Open https://platform.openai.com/api-keys to create an API key.
- Paste the key into the `.env` file

```python
import os
from dotenv import load_dotenv
from openai import OpenAI
```

```python

# load_dotenv will define all variables from the .env file as environment variables
load_dotenv(override=True)

# Here we read the API key from the environment variable
api_key = os.getenv('OPENAI_API_KEY')

if not api_key:
    print("No API key defined in the .env file! Please add your OpenAI API key to the .env file")
elif not api_key.startswith("sk-proj-"):
    print("The API key has the wrong format - it should start with 'sk-proj-'")
else:
    print("API key was defined with the correct format!")
```

```python
client = OpenAI()

prompt = "This is my first programmatic interaction with an LLM - I'm so excited!"

response = client.responses.create(
    model="gpt-5.4-nano", input=prompt
)

print(response.output_text)
```
