# Your First Local LLM Call

In the previous notebook, you called `gpt-5.4-nano` through the OpenAI API.

That is the simplest way to get started, but every API call uses paid credits. In this notebook, you will use Ollama to run an open-source model locally on your own machine.

Local models are usually less powerful than the strongest hosted models, but they are useful for many practical tasks such as summarization, translation, and classification.

## Install Ollama

Download and install Ollama from [ollama.com](https://ollama.com/).

After installation, open a new terminal and run:

`ollama --version`

If a version number appears, Ollama is installed correctly.

## Choose a Local Model

Ollama can run many different local models. You can browse them at [ollama.com/search](https://ollama.com/search).

In this lesson, we use Google's Gemma 4 family. Model variants with more parameters are usually more capable, but they also need more memory.

If your machine has enough memory (16GB+), use the 4-billion-parameter Gemma 4 variant:

`ollama pull gemma4:e4b`

If that model is too large for your machine, use a smaller model instead, for example:

`ollama pull gemma3:1b`

Keep the model name you choose in mind, because you will use the same name in the Python code below.

## Try the Model in the Terminal

Before using Python, first check that the model works in the terminal:

`ollama run gemma4:e4b`

Send a short message like `Who are you?`.

When you are done, type `/bye` to exit the chat.

If you downloaded a different model, replace `gemma4:e4b` with that model name.

## Connect to Ollama

Ollama runs a local API server on port `11434`.

The `api_key` value is required by the OpenAI SDK, but Ollama ignores it. That is why we can use a simple placeholder value.

If you chose a different model above, update `MODEL` before running the next cells.

```python
from openai import OpenAI

OLLAMA_BASE_URL = "http://localhost:11434/v1"
MODEL = "gemma4:e4b"

client = OpenAI(
    base_url=OLLAMA_BASE_URL,
    api_key="ollama",
)
```

## Send a Request

This request looks almost the same as the OpenAI API request from the previous notebook.

The main difference is the `base_url` you used when creating the client. The request now goes to Ollama on your machine instead of to OpenAI's servers.

```python
prompt = "Who are you?"

response = client.responses.create(model=MODEL, input=prompt)

print(response.output_text)
```

If this cell fails, check these things first:

1. Ollama is installed and running.
2. The model was downloaded with `ollama pull gemma4:e4b`.
3. `MODEL` matches the model name you downloaded.
4. If your machine does not have enough memory for Gemma 4, switch to a smaller model such as `gemma3:1b`.
