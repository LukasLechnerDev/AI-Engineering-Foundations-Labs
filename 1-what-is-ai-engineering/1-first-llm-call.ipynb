# Your First LLM Call

In this notebook, you will make your first call to the OpenAI API.

Before you run the code, add your OpenAI API key to the `.env` file.

- Run `cp .env.example .env` in the terminal.
- Create an API key at [platform.openai.com/api-keys](https://platform.openai.com/api-keys).
- Paste the key into `.env`.

Reference: [OpenAI Developer Quickstart](https://developers.openai.com/api/docs/quickstart)

```python
import os
from dotenv import load_dotenv
from openai import OpenAI
```

## Check Your API Key

This cell loads the variables from `.env` and checks whether `OPENAI_API_KEY` is present and has the expected format.

```python
# load_dotenv will define all variables from the .env file as environment variables
load_dotenv(override=True)

# Here we read the API key from the environment variable
api_key = os.getenv("OPENAI_API_KEY")

if not api_key:
    print(
        "No API key defined in the .env file! Please add your OpenAI API key to the .env file"
    )
elif not api_key.startswith("sk-proj-"):
    print("The API key has the wrong format - it should start with 'sk-proj-'")
else:
    print("API key was defined with the correct format!")
```

## Send Your First Request

In the final cell, you create an `OpenAI` client, define a short prompt, send it to the model, and print the response.

```python
client = OpenAI()

prompt = "This is my first programmatic interaction with an LLM - I'm so excited!"

response = client.responses.create(model="gpt-5.4-nano", input=prompt)

print(response.output_text)
```
