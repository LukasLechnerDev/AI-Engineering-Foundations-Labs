# Development setup

Great, now that you have cloned the repo, it's time to set up your development environment so that you can execute cells in Jupyter notebooks (`*.ipynb` files) and Python modules (`*.py` files) in this project.

## Visual Studio Code Extensions

You need to install 2 extensions: 
1. The "Python" extension from Microsoft
2. The "Jupyter" extension from Microsoft

## UV

### Installation

In this repository, we are using `uv`, which is a modern, popular, and fast package manager for Python. 

The installation of `uv` should be pretty straightforward: https://docs.astral.sh/uv/

The easiest way is to open up a new terminal in Visual Studio Code and run the commands there.

You probably need to start a new terminal session after installing `uv`. 

Then you can execute `uv --version` to verify your installation

If you already have `uv` installed, it doesn't hurt to update it to the latest version with `uv self update`

### Creating the virtual environment and installing the project dependencies

Now that you have `uv` installed, it's time to run `uv sync` in your terminal. 

This does the following: 
1. A new virtual environment is created. You now have a `.venv` folder in your project root. This virtual environment is a small isolated Python setup just for this project.
2. All project dependencies that are defined in `pyproject.toml` are installed in that new virtual environment. 

If the following popup appears, click on "Yes" ![image 2.png](<images/venv-popup.png>)

This sets the Python interpreter of the new virtual environment for the whole project.

## Executing Notebooks

Now you should be able to run cells in a Jupyter notebook, like this file. 

For each notebook, you first need to define the kernel on which you want to run it. 

To do that, click on "Select Kernel" on the top right, and then select the one that references the `.venv` folder of our project. It is usually highlighted with a star in VS Code.

After you have selected the kernel, try to execute the next cell. You can either execute a code cell by clicking on the "play" button to the left of the cell, or by clicking into the cell and pressing "Shift" and "Return". 
It should successfully print out the result of the complex calculation.

```python
result = 1 + 1

print(result)
```

If you have problems executing the cell, maybe ChatGPT or another AI tool can help you. If you are still stuck, don't hesitate to shoot me a question via Udemy!

Maybe you wonder why we didn't have to install Python on our machine. The reason is that `uv` itself installs a version of Python when it creates the virtual environment. This Python version is located under `.venv/bin/python`.

The version that should be used for the virtual environment can be defined in the `.python-version` file.

### Running ordinary Python modules

Okay, so now that you can run notebook cells, let's check if you are also able to run ordinary Python modules. To do that, let's open up the file `0-uv-check.py` in the `1-what-is-ai-engineering` folder. That's the file we are now going to execute. 
The easiest way to do so is to open up a new terminal and execute the following command: `uv run python 1-what-is-ai-engineering/0-uv-check.py`

Alternatively, you can also execute this file by clicking on the Play button at the top of VS Code. 

The short Python code should be executed, and the version of the OpenAI client library should be printed out. If that doesn't work, you will probably get the error: 
> ModuleNotFoundError: No module named 'openai'

To fix this, you need to set the correct Python interpreter by entering a ">" in the top search field, then selecting "Python: Select Interpreter", and using the one from our `.venv` folder.
