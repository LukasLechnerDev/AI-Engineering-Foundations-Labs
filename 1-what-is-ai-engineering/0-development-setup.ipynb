## Visual Studio Code Extensions

Install these two VS Code extensions:

1. `Python` by Microsoft
2. `Jupyter` by Microsoft

## Install `uv`

This repository uses `uv`, a fast modern Python package manager.

Installation instructions: [docs.astral.sh/uv](https://docs.astral.sh/uv/)

The easiest way is to open a terminal in VS Code and run the install command there.

After installation, you will usually need to open a new terminal session.

Then run `uv --version` to verify that it works.

If you already have `uv` installed, you can update it with `uv self update`.

## Run `uv sync`

Once `uv` is installed, run `uv sync` in the terminal.

This does two important things:

1. It creates a virtual environment in the `.venv` folder.
2. It installs the project dependencies from `pyproject.toml` into that environment.

If VS Code shows the following popup, click `Yes`: ![image 2.png](<images/venv-popup.png>)

That tells VS Code to use the Python interpreter from `.venv` for this project.

## Run Notebook Cells

You should now be able to run cells in this notebook.

For each notebook, you first need to select the kernel.

Click `Select Kernel` in the top right and choose the interpreter that points to the project's `.venv` folder. In VS Code, this option is often marked with a star.

After that, run the next code cell.

You can run a cell by clicking the play button on the left or by pressing `Shift + Enter` inside the cell.

### Quick Notebook Check

```python
result = 1 + 1

print(result)
```

If this cell does not run, the most likely issue is that the wrong kernel is selected.

First check whether the notebook is using the `.venv` interpreter. If you are still stuck, an AI tool or the course Q&A can help you troubleshoot the setup.

## Why You Do Not Need to Install Python Separately

You might wonder why we did not install Python manually.

The reason is that `uv` can manage the Python version for the project and create the virtual environment for you. The interpreter used by this project lives inside `.venv`.

The target Python version is defined in `.python-version`.

## Run a Regular Python File

Now that notebook cells work, the next step is to check whether regular Python files work too.

Open `1-what-is-ai-engineering/0-uv-check.py` and run:

`uv run python 1-what-is-ai-engineering/0-uv-check.py`

This should print the installed version of the OpenAI client library.

You can also try running the file with the play button in VS Code.

If that does not work, you may see this error:
> ModuleNotFoundError: No module named 'openai'

To fix it, open the VS Code command palette, run `Python: Select Interpreter`, and choose the interpreter from the `.venv` folder.
