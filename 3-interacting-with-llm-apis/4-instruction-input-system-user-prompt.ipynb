```python
from dotenv import load_dotenv
from IPython.display import Markdown, display
from openai import OpenAI

load_dotenv(override=True)

client = OpenAI()
```


### "New" Syntax in new Responses API


The instructions more or less define the task *description*, and the input defines the task *data*. In other words, you could say that the instructions define stable rules, and the input defines the variable input.

| API part | What goes there | Example |
| --- | --- | --- |
| `instructions` | Stable app rules and behavior | "You are an AI Engineer Job Advisor." |
| `input` | Changing task data or user request | "Tell me the #1 skill needed to become an AI Engineer." |

```python
response = client.responses.create(
    model="gpt-5.4-nano",
    instructions="""You are a AI Engineer Job Advisor that provides career advice 
    to people interested in becoming AI Engineers.""",
    input="Tell me the #1 skill needed to become an AI Engineer.",
)

display(Markdown(response.output_text))
```

### "Classic" Syntax in old Chat Completions API

The *new* API, called `ResponsesAPI` was introduced in March 2025, and is the recommended way of interacting with the OpenAI API https://openai.com/index/new-tools-for-building-agents/.

"Instructions" and "Input" are still commonly called "System Prompt" and "User Prompt". This is because calls in the *old* API, the Chat Completions API looked like this: 

```python
response = client.chat.completions.create(
    model="gpt-5.4-nano",
    messages=[
        {
            "role": "system",
            "content": """You are a AI Engineer Job Advisor that provides career advice 
                        to people interested in becoming AI Engineers.""",
        },
        {
            "role": "user",
            "content": "Tell me the #1 skill needed to become an AI Engineer.",
        },
    ],
)

display(Markdown(response.choices[0].message.content))
```

### "Classic" Syntax in new Responses API

The Responses API can also accept a list of messages. In this format, the app-level instructions use the `developer` role, and the user's request uses the `user` role.

```python
response = client.responses.create(
    model="gpt-5.4-nano",
    input=[
        {
            "role": "developer",
            "content": """
                        You are a AI Engineer Job Advisor that provides career advice 
                        to people interested in becoming AI Engineers.
                        """,
        },
        {
            "role": "user",
            "content": "Tell me the #1 skill needed to become an AI Engineer.",
        },
    ],
)

display(Markdown(response.output_text))
```
