```python
import math

import pandas as pd

from dotenv import load_dotenv
from openai import OpenAI

load_dotenv(override=True)

client = OpenAI()

prompt = "The best programming language for AI engineering is"

next_token_response = client.responses.create(
    model="gpt-5.4-mini",
    input=prompt,
    max_output_tokens=1,
    top_logprobs=10,
    include=["message.output_text.logprobs"],
)

output_text = next_token_response.output[0].content[0]
next_token = output_text.logprobs[0]

rows = []
for token_logprob in next_token.top_logprobs:
    probability = math.exp(token_logprob.logprob)
    rows.append({
        "token": repr(token_logprob.token),
        "probability": probability,
        "percentage": f"{probability:.2%}",
        "logprob": token_logprob.logprob,
    })

print(f"Prompt: {prompt!r}")
print(f"Sampled next token: {next_token.token!r}")
print(
    "The table shows the top token candidates returned by the API, not the full vocabulary."
)
display(pd.DataFrame(rows))
```
