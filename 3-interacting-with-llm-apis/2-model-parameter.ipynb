# The `model` parameter

Before we look at new parameters, let's first go a bit deeper into one parameter we already use all the time: `model`.

The `model` parameter determines which exact model should process the request.

OpenAI Responses API reference: https://developers.openai.com/api/reference/resources/responses/methods/create

OpenAI models overview: https://developers.openai.com/api/docs/models

At the time of this recording, OpenAI's latest model is GPT-5.5.

A useful mental model is:

> If you are not sure where to start, start with the most powerful model. Later, when we add evaluations, we can test whether a cheaper or faster model gives us the same quality.

In the course repository, we use a cheaper model so that the API calls do not get too expensive.

## Frontier models

Frontier models are the most capable models currently available. At the time of this recording, the recent frontier models are GPT-5.5, GPT-5.4, and GPT-5.4-mini.

## GPT-5.4

GPT-5.4 is the strongest option. Use it for hard, ambiguous, important tasks where quality matters most.

It has the best reasoning, but it is slower than the smaller variants and more expensive.

## GPT-5.4-mini

GPT-5.4-mini is the smaller, faster, and cheaper version.

It is usually enough for everyday tasks like classification, extraction, and summarization.

That makes it a good option for our course repository, because these are the tasks we mainly do.

## GPT-5.4-nano

GPT-5.4-nano is the cheapest and fastest version.

Use it for very simple, high-volume tasks like tagging, routing, and ranking.
